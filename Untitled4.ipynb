```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#line chart
x=[1,2,3,4,5]
y=[1,4,9,16,25]
#plt.plot(x,y)
plt.figure(figsize=(7,5))
plt.plot(x,y)
plt.title('relation btw x and y')
plt.xlabel('number')
plt.ylabel('square')
plt.tight_layout()
plt.show()


```

```python
subjects=["maths",'physics','cs','english']
score=[88,76,76,90]
plt.bar(subjects,score,color='red',edgecolor='black')
plt.title('bar chart')
plt.xlabel('subjects')
plt.ylabel('score')
plt.tight_layout()
plt.show()
```

```python
hours = np.random.randint(1,10,50);
marks = hours*8 + np.random.randint(1,20,50)
plt.scatter(hours,marks, color='red')
plt.plot(hours,marks)
plt.show()
```

```python
data=np.random.normal(60,10,50)
plt.hist(data,bins=10,color='red')
plt.title('histogram')
plt.xlabel('marks')
plt.ylabel('frequency')
plt.tight_layout()
plt.show()

```

```python
subjects=["maths",'physics','cs','english']
score=[30,40,10,20]
plt.pie(score,labels=subjects,autopct='%.2f%%',shadow=True)

plt.title('pie chart')
plt.show()

```

```python
batch1 =np.random.randint(10,70,100)
plt.boxplot(batch1,)
plt.tight_layout
plt.show()


```
